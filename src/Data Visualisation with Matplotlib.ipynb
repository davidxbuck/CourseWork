```python
from sklearn import datasets
iris = datasets.load_iris()
# Location for legend
# 'upper left' 2 
# 'upper center' 9 
# 'upper right' 1
# 'center left' 6 
# 'center' 10 
# 'center right' 7
# 'lower left' 3
# 'lower center' 8 
# 'lower right' 4
# 'best' 0 'right' 5

colors = ['red', 'green', 'blue']
measures = [0, 1]
for type in list(set(iris.target)):
    flower = iris.data[np.equal(type, iris.target),:]
    plt.scatter(flower[:,measures[0]], flower[:,measures[1]], marker='o', color=colors[type], label=iris.target_names[type])
plt.legend(loc='upper right')
plt.title('Iris data')
plt.xlabel(iris.feature_names[measures[0]])
plt.ylabel(iris.feature_names[measures[0]])

# Options for annotate
# s          text of label
# xy         coordinates to annotate
# xytext     coordinates of label
# arrowprops controls drawing of arrow

# plt.annotate('setosa', xy=(5.0, 3.5))     # Simple text annotation, no arrow
# plt.annotate('virginica', xy=(7.25, 3.5))
# plt.annotate('versicolor', xy=(5.0, 2.0))

plt.annotate('setosa', xy=(5.0, 3.5), xytext=(4.25, 4.0), arrowprops={'color':'red'})
plt.annotate('virginica', xy=(7.2, 3.6), xytext=(6.5, 4.0), arrowprops={'color':'blue'})
plt.annotate('versicolor', xy=(5.05, 1.95), xytext=(5.5, 1.75), arrowprops={'color':'green'})
plt.style.use('fivethirtyeight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()

# Save the figure to 'sine_mesh.png'
plt.savefig('./output/sine_mesh.png')
```

## Histogramming


```python
from random import randint

x = np.array([randint(0, 64) for _ in range(8192)])
y = np.array([randint(0, 64) for _ in range(8192)])

plt.hist2d(x,y, bins=(20, 20), range=((30, 50), (8, 48)), cmap='viridis') # 20x20 bins, x, y restricted

# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('hist2d() plot')
plt.show()
```

```python
# Generate a 2d histogram with hexagonal bins
plt.hexbin(x, y, gridsize=(15, 12), extent=(30, 50, 8, 48), cmap='viridis')
        
# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('hexbin() plot')
plt.show()
```

# Image manipulation



```python
orig = plt.imread('../input/moon-low-contrast.png')
pixels = orig.flatten()
plt.hist(pixels, bins=256, range=(0,1), normed=True, color='blue', alpha=0.3)
plt.show()   # alpha makes the histogram partiallly tranparent

plt.imshow(orig)
plt.axis('off')
plt.show()
```

```python
origshape = orig.shape
minval, maxval = orig.min(), orig.max()
print(minval, maxval)
rescaled = (1/(maxval-minval)) * (pixels - minval)
print(rescaled.min(), rescaled.max())
plt.hist(pixels, bins=256, range=(0,1), normed=True, color='blue', alpha=0.3)
plt.hist(rescaled, bins=256, range=(0,1), normed=True, color='red', alpha=0.3)
plt.legend(['original', 'rescaled'])
plt.show()

rescaledimg = rescaled.reshape(origshape[0], -1)
plt.imshow(rescaledimg)
plt.axis('off')
plt.show()
plt.savefig('../output/modified-moon.png')
```

```python
#   CDF

plt.hist(pixels, bins=256, range=(0,1), normed=True, color='blue', alpha=0.3)
plt.twinx()
orig_cdf, bins, patches = plt.hist(pixels, cumulative=True, bins=256, range=(0,1), normed=True, color='red', alpha=0.3)
plt.title('Image histogram and CDF')
plt.xlim((0, 1))
plt.show()

plt.hist(rescaled, bins=256, range=(0,1), normed=True, color='blue', alpha=0.3)
plt.twinx()
rescaled_cdf, bins, patches = plt.hist(rescaled, cumulative=True, bins=256, range=(0,1), normed=True, color='red', alpha=0.3)
plt.title('Rescaled Image histogram and CDF')
plt.xlim((0, 1))
plt.show()
```

```python
import numpy as np
new_pixels = np.interp(pixels, bins[:-1], orig_cdf*1)
new = new_pixels.reshape(orig.shape)
plt.imshow(new)
plt.axis('off')
plt.title('Equalized image')
plt.show()
```
