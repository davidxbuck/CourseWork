```python
import matplotlib.pyplot as plt
from random import randint
year = list(range(1995, 2011))
values = [randint(x, 20+x) for x in range(5, 21)]
value2 = [randint(x, 20+x) for x in range(5, 21)]
```

```python
plt.plot(year, values, color='red')
plt.plot(year, value2, color='blue')
plt.show()
```

```python
# Syntax: 
#
# plt.axes( [x_lo, y_lo, width, height] ) 
plt.axes([0.05, 0.05, 0.425, 0.9])
plt.plot(year, values, color='red')

plt.axes([0.525, 0.05, 0.425, 0.9])
plt.plot(year, value2, color='blue')
plt.show()
```

```python
# Syntax: 
# subplot(nrows, ncols, nsubplot) 
#
# Subplot ordering:
# - Row-wise from top left"
# - Indexed from 1
plt.subplot(2,1,1)
plt.plot(year, values, color='red')

plt.subplot(2,1,2)
plt.plot(year, values, color='red')

plt.tight_layout()
plt.show()
```

```python
# Syntax: 
# axis([xmin, xmax, ymin, ymax]) # Equivalent to xlim+ylim
# xlim([xmin, xmax])
# ylim([ymin, ymax])
# xlabel("string")
# ylabel("string")
# title("string")

# axis(off) turns off axis lines, labels
# axis(equal) equal scaling on x, y axes
# axis(square) forces square plot
# axis(tight) sets xlim(), ylim() to show all data

plt.plot(year, values, color='red')
plt.plot(year, values, color='red')
plt.xlim((2000,2005))
plt.title("Title of the Graph")
plt.xlabel("Year")
plt.ylabel("Arbitrary Units")
plt.plot()
plt.savefig('output/xlim_and_ylim.png')
```

```python
from sklearn import datasets
iris = datasets.load_iris()
# Location for legend
# 'upper left' 2 
# 'upper center' 9 
# 'upper right' 1
# 'center left' 6 
# 'center' 10 
# 'center right' 7
# 'lower left' 3
# 'lower center' 8 
# 'lower right' 4
# 'best' 0 'right' 5

colors = ['red', 'green', 'blue']
measures = [0, 1]
for type in list(set(iris.target)):
    flower = iris.data[np.equal(type, iris.target),:]
    plt.scatter(flower[:,measures[0]], flower[:,measures[1]], marker='o', color=colors[type], label=iris.target_names[type])
plt.legend(loc='upper right')
plt.title('Iris data')
plt.xlabel(iris.feature_names[measures[0]])
plt.ylabel(iris.feature_names[measures[0]])

# Options for annotate
# s          text of label
# xy         coordinates to annotate
# xytext     coordinates of label
# arrowprops controls drawing of arrow

# plt.annotate('setosa', xy=(5.0, 3.5))     # Simple text annotation, no arrow
# plt.annotate('virginica', xy=(7.25, 3.5))
# plt.annotate('versicolor', xy=(5.0, 2.0))

plt.annotate('setosa', xy=(5.0, 3.5), xytext=(4.25, 4.0), arrowprops={'color':'red'})
plt.annotate('virginica', xy=(7.2, 3.6), xytext=(6.5, 4.0), arrowprops={'color':'blue'})
plt.annotate('versicolor', xy=(5.05, 1.95), xytext=(5.5, 1.75), arrowprops={'color':'green'})
plt.style.use('fivethirtyeight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()

# Save the figure to 'sine_mesh.png'
plt.savefig('./output/sine_mesh.png')
```

## Histogramming


```python
from random import randint

x = np.array([randint(0, 64) for _ in range(8192)])
y = np.array([randint(0, 64) for _ in range(8192)])

plt.hist2d(x,y, bins=(20, 20), range=((30, 50), (8, 48)), cmap='viridis') # 20x20 bins, x, y restricted

# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('hist2d() plot')
plt.show()
```

```python
# Generate a 2d histogram with hexagonal bins
plt.hexbin(x, y, gridsize=(15, 12), extent=(30, 50, 8, 48), cmap='viridis')
        
# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('hexbin() plot')
plt.show()
```
