```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips') # imports https://github.com/mwaskom/seaborn-data/tips.csv as df
sns.lmplot(x= 'total_bill', y='tip', data=tips)

plt.show()
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1')
plt.show()

```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
plt.show()
```

```python
sns.residplot(x='total_bill', y='tip',data=tips,color='indianred')
plt.show()
```

```python
auto = pd.read_csv('./input/auto-mpg.csv')
auto.head()

# Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')

# Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, color='blue', scatter=None, order=1, label='order 1')

# Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, color='green', scatter=None, order=2, label='order 2')

# Add a legend and display the plot
plt.legend(loc='upper right')
plt.show()

```

```python
# Plot a linear regression between 'weight' and 'hp', with a hue of 'origin' and palette of 'Set1'
sns.lmplot(x='weight', y='mpg', data=auto, hue='origin', palette='Set1')

# Display the plot
plt.show()
```

```python
# Plot linear regressions between 'weight' and 'mpg' grouped row-wise by 'origin'
sns.lmplot(x='weight', y='mpg', data=auto, row='origin' )

# Display the plot
plt.show()
```

```python
sns.stripplot(y= 'tip', data=tips, jitter=False) # default jitter=True
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(y= 'tip', data=tips, jitter=True) # default jitter=True
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```

```python
# Generate a swarm plot of 'hp' grouped horizontally by 'cylinders'  
plt.subplot(2,1,1)
sns.swarmplot(x='cylinders', y='hp', data=auto)

# Generate a swarm plot of 'hp' grouped vertically by 'cylinders' with a hue of 'origin'
plt.subplot(2,1,2)
sns.swarmplot(x='hp', y='cylinders', data=auto, hue='origin', orient='h', size=3)

# Display the plot
plt.show()
```

```python
#   Combining plots

sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

###   Bivariate Plots


```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# kind='scatter' uses a scatter plot of the data points
# kind='reg' uses a regression plot (default order 1)
# kind='resid' uses a residual plot
# kind='kde' uses a kernel density estimate of the joint distribution
# kind='hex' uses a hexbin plot of the joint distribution
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', size=3)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, kind='reg', hue='sex', height=3)
plt.show()

```
